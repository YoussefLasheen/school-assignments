```python
import numpy as np
```

```python
price_unit = np.array([101,	90,	99	,123,	111,	98	,109	,104	,91	,122,	116,	109])
demand = np.array([1005	,1154,	1033	,777	,876	,1046,	904	,958,	1144,	792,	828,	913])
sse = np.array([])


rp = 100
def calculate_demand(rd, price, rp, e):
  return rd*(price/rp)**e

for rd in range(500,1500,10):
  for e_base in range(-2,-1):
    e = e_base * 0.01
    for i in range(12):
      sse = np.append(sse, (calculate_demand(rd, price_unit[i], 100, e) - demand[i])**2)

print(sse)

```
